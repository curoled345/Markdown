## 1. 차선 감지 (기초 이미지 처리)

차선을 감지하기 위해 OpenCV를 사용하여 이미지를 흑백으로 변환하고 가장자리를 감지합니다.

```python
import cv2
import matplotlib.pyplot as plt

# 이미지 불러오기
# Replace "/path/to/your/image.jpg" with the actual path to your image file
img = cv2.imread("/path/to/your/image.jpg")

if img is None:
    print("Error: Could not read the image. Please check the file path.")
else:
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)  # 흑백 변환
    edges = cv2.Canny(gray, 100, 200)  # 가장자리 감지

    plt.imshow(edges, cmap='gray')
    plt.title("Edge Detection")
    plt.axis("off")
    plt.show()
```

## 2. 속도 제어 시뮬레이션

목표 속도에 도달할 때까지 속도를 증가시키는 간단한 시뮬레이션입니다.

```python
current_speed = 0
target_speed = 60

while current_speed < target_speed:
    current_speed += 5
    print("현재 속도:", current_speed, "km/h")
```

## 3. 거리 센서 데이터 처리

센서에서 입력된 거리 데이터를 기반으로 충돌 여부를 판단합니다.

```python
distance = 20  # cm 단위 거리
if distance < 30:
    print("장애물 감지! 멈춥니다.")
else:
    print("전진합니다.")
```

## 4. GPS 좌표 기반 위치 판별

현재 GPS 좌표가 특정 영역(예: 서울시 경계)에 있는지 확인합니다.

```python
latitude = 37.55
longitude = 126.98

if 37.4 < latitude < 37.7 and 126.8 < longitude < 127.2:
    print("서울시 안에 있습니다.")
else:
    print("서울시 밖입니다.")
```

## 5. 신호등 인식 시뮬레이션

신호등 색상에 따라 차량의 행동을 결정하는 시뮬레이션입니다.

```python
signal = "red"

if signal == "red":
    print("정지합니다.")
elif signal == "yellow":
    print("감속합니다.")
elif signal == "green":
    print("출발합니다.")
```

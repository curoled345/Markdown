## 조건문

조건문은 코드의 흐름을 제어할 때 사용 if, elif, else 키워드

```python
x = 10
y = 20
if x < y:
    print("x는 y보다 작습니다.")
elif x == y:
    print("x와 y는 같습니다.")
else:
    print("x는 y보다 큽니다.")
```

## 반복문

반복문 : for문과 while문  / 리스트나 범위를 반복할 수 있음

```python
for i in range(5):
    print(i)

i = 0
while i < 5:
    print(i)
    i += 1
```

## 함수

함수는 특정 작업을 수행하는 코드 블록, `def` 키워드로 정의

```python
def greet(name):
    print("Hello, " + name)

greet("Alice")
```

## 클래스

클래스는 객체를 생성하기 위한 템플릿, 객체 지향 프로그래밍

```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def greet(self):
        print("Hi, my name is " + self.name)

p1 = Person("John", 36)
p1.greet()
```

## 파일 처리

with 문을 사용하면 파일을 자동 닫음

```python
with open("example.txt", "w") as f:
    f.write("Hello file!")

with open("example.txt", "r") as f:
    print(f.read())
```

## 예외 처리

try, except 블록을 사용해 예외를 처리 가능

```python
try:
    print(10 / 0)
except ZeroDivisionError:
    print("0으로 나눌 수 없습니다.")
finally:
    print("예외 처리가 완료되었습니다.")
```

## 문자열 (Strings)

문자열은 텍스트 데이터를 저장할 수 있으며 다양한 내장 메서드를 제공

```python
txt = " Hello, Python! "
print(txt.strip())       # 앞뒤 공백 제거
print(txt.lower())        # 소문자 변환
print(txt.upper())        # 대문자 변환
print(txt.replace("Python", "World"))
print("Python" in txt)
```

## 날짜 처리

Python에서는 datetime 모듈을 사용하여 날짜 및 시간 데이터를 처리

```python
import datetime
now = datetime.datetime.now()
print("현재 시간:", now)
print("연도:", now.year)
print("월:", now.month)
```

## 반복문 내 리스트 컴프리헨션

리스트 컴프리헨션은 간결하게 리스트를 생성할 수 있는 문법

```python
squares = [x**2 for x in range(10)]
print(squares)
```

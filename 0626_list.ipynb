```python
import random
import matplotlib.pyplot as plt
import csv

true_distance = 10.0  # 실제 거리

noise_ranges = {"lidar": 0.3, "radar": 0.5, "camera": 0.8}
weights = {"lidar": 0.5, "radar": 0.3, "camera": 0.2}
threshold = 0.5  # 허용 오차 범위


def noisy_measurement(true_value, noise_range):
    """센서 노이즈 포함 측정값 생성"""
    return round(true_value + random.uniform(-noise_range, noise_range), 2)

def sensor_fusion(lidar, radar, camera):
    """단순 평균과 가중 평균 계산"""
    simple_avg = round((lidar + radar + camera) / 3, 2)
    weighted_avg = round(
        lidar * weights["lidar"] +
        radar * weights["radar"] +
        camera * weights["camera"], 2
    )
    return simple_avg, weighted_avg

def save_to_csv(results):
    """결과를 CSV로 저장"""
    with open("sensor_results.csv", mode="w", newline="", encoding="utf-8") as file:
        writer = csv.writer(file)
        writer.writerow(["횟수", "LiDAR", "Radar", "Camera", "단순평균", "가중평균", "오차", "상태"])
        for row in results:
            writer.writerow(row)



results = []

lidar_list, radar_list, camera_list, fusion_list = [], [], [], []

for i in range(1, 11):  # for문
    lidar = noisy_measurement(true_distance, noise_ranges["lidar"])
    radar = noisy_measurement(true_distance, noise_ranges["radar"])
    camera = noisy_measurement(true_distance, noise_ranges["camera"])

    simple_avg, weighted_avg = sensor_fusion(lidar, radar, camera)

    error = round(abs(true_distance - weighted_avg), 2)
    status = "정상" if error <= threshold else "경고"

    results.append([i, lidar, radar, camera, simple_avg, weighted_avg, error, status])
    lidar_list.append(lidar)
    radar_list.append(radar)
    camera_list.append(camera)
    fusion_list.append(weighted_avg)

plt.figure(figsize=(12, 6))
plt.plot(lidar_list, label="LiDAR", marker="o")
plt.plot(radar_list, label="Radar", marker="s")
plt.plot(camera_list, label="Camera", marker="^")
plt.plot(fusion_list, label="Weighted Avg", linestyle="--", color="black", linewidth=2)
plt.axhline(y=true_distance, color="gray", linestyle=":", label="Ground Truth")

plt.title(" 자율주행 센서 통합 거리 측정 결과")
plt.xlabel("측정 회차")
plt.ylabel("거리 (m)")
plt.legend()
plt.grid()
plt.tight_layout()
plt.show()

save_to_csv(results)
```
